# Importando biblioteca pandas e definindo o caminho do dataset

```python
import pandas as pd
path = r"Global YouTube Statistics.csv"
path
```

# Enxergando tabela do dataset

```python
tabela_inicial = pd.read_csv(path, encoding='latin-1') #Parametros para a leitura apropriada da tabela
display(tabela_inicial) #exibindo a tabela abaixo
```

# Buscando qual canal tem a maior quantidade de inscritos
#### Sabendo dessa informação, podemos aferir quais os tipos de canal mais populares, ou seja, com maior audiência
# Depois, buscando qual categoria possui a maior quantidade de videos
#### Tendo em mente a quantidade de inscritos em cada categoria, sabendo a quantidade de uploads é possivel julgar o quão atrativo um upload é em cada categoria.
# Com essas duas informações, é calculado quantos inscritos - em média - cada upload rende
#### Uma média juntando as informações obtidas nas duas perguntas anteriores, tornando uma resposta para essas questões mais fácil de ser enxergada

```python
tabela_tipo_de_canal = tabela_inicial.groupby('category')[['subscribers', 'uploads']].sum() #criando a tabela abaixo com as informações da tabela inicial
tabela_tipo_de_canal['subscribers per upload'] = tabela_tipo_de_canal['subscribers']//tabela_tipo_de_canal['uploads'] #criando a coluna de inscritos por upload
tabela_tipo_de_canal = tabela_tipo_de_canal.sort_values(by="subscribers per upload",ascending = False) #organizando em ordem decrescente
display(tabela_tipo_de_canal) #exibindo a tabela
```

# Gerando um gráfico para melhor visualizar as informações

```python
import plotly.express as px
grafico = px.bar(tabela_tipo_de_canal, x = tabela_tipo_de_canal.index, y = 'subscribers per upload')
grafico.show()
```

# Quantos canais existem de cada tipo? 

#### Essa informação é pertinente para a conclusão porque nos mostra o nivel de concorrência e o quão nichado são os possiveis tipos de canal

```python
tabela_Canais_por_tipo = tabela_inicial.groupby('category')[['category']].count()

tabela_Canais_por_tipo = tabela_Canais_por_tipo.rename(columns = {'category':'channel count'})

tabela_Canais_por_tipo = tabela_Canais_por_tipo.sort_values(by = 'channel count', ascending = False)

display(tabela_Canais_por_tipo)
```

# Uma tabela feita para encontrar o pais com maior visibilidade de comédia
#### Sabendo disso nos ajuda a direcionar melhor o conteúdo do canal de forma geral

```python
tabela_filtrada = tabela_inicial[tabela_inicial['category'] == 'Comedy']
tabela_auxiliar = tabela_filtrada[["Abbreviation", "category"]].groupby(["category", "Abbreviation"])[['Abbreviation']].count() # criando uma tabela com a contagem de quantos canais existem por pais
tabela_auxiliar = tabela_auxiliar.rename(columns = {'Abbreviation':'Abbreviation_count'}) # mudando o nome da segunda coluna para melhor compreensão
tabela_auxiliar = tabela_auxiliar.sort_values(by='Abbreviation_count', ascending=False)
display(tabela_auxiliar)
```
